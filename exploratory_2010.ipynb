#Applying the composition rates to recycling

```python
hhkerb_rec_ton_drs = hhkerb_rec_ton_agg
```

```python
hhkerb_rec_ton_drs.head()
```

```python
hhkerb_rec_ton_drs['DRS Ferrous Cans'] = hhkerb_rec_ton_drs['Steel cans']*.18
hhkerb_rec_ton_drs['DRS Aluminium Cans'] = hhkerb_rec_ton_drs['Aluminium cans']*.80
hhkerb_rec_ton_drs.head()
```

```python
#Using WRAP-calculated rates
hhkerb_rec_ton_drs['DRS Glass Bottles'] = hhkerb_rec_ton_drs['Mixed glass']*.66
```

```python
#Mostly, 'mixed Plastic Bottles' are treated as PET, HDPE, and Other. So DRS is assumed to be only PET and HDPE, 96%
hhkerb_rec_ton_drs['DRS Plastic Bottles'] = hhkerb_rec_ton_drs['Mixed Plastic Bottles']*.96
hhkerb_rec_ton_drs[['Mixed Plastic Bottles','DRS Plastic Bottles']].head(22)
```

```python
# If info not provided from Mixed Plastic Bottles, use Plastics. For most of these LAs, plastics are dense plastics. 
hhkerb_rec_ton_drs['DRS Plastic Bottles'] = np.where(hhkerb_rec_ton_drs['DRS Plastic Bottles'].isnull(), 
                                                     hhkerb_rec_ton_drs['Plastics']*.68, 
                                                     hhkerb_rec_ton_drs['DRS Plastic Bottles'])
hhkerb_rec_ton_drs[['Mixed Plastic Bottles','Plastics', 'DRS Plastic Bottles']].head(22)
```

```python
#For Swansea, Plastics are dense plastics plus plastic film, so the rate is 57% (from dividing 6.69 from 11.66)
hhkerb_rec_ton_drs[hhkerb_rec_ton_drs['Authority'] == 'City  and County of Swansea ']
hhkerb_rec_ton_drs['DRS Plastic Bottles'] = np.where(hhkerb_rec_ton_drs['Authority'] == 'City  and County of Swansea ',
                                                     hhkerb_rec_ton_drs['Plastics']*.57, 
                                                     hhkerb_rec_ton_drs['DRS Plastic Bottles'])
hhkerb_rec_ton_drs[['Authority', 'Mixed Plastic Bottles','Plastics', 'DRS Plastic Bottles']].head(22)
```

```python
hhkerb_rec_ton_drs.head(22)
```

```python
# Use Mixed cans if they are available (.27 is aluminium, .73 is ferrous) ... except for Neath Port Talbot and Powys
hhkerb_rec_ton_drs['DRS Aluminium Cans'] = np.where(hhkerb_rec_ton_drs['Mixed cans'].notnull(), 
                                                    hhkerb_rec_ton_drs['Mixed cans']*.27*.8, 
                                                    hhkerb_rec_ton_drs['DRS Aluminium Cans'])
hhkerb_rec_ton_drs['DRS Aluminium Cans'] = np.where(hhkerb_rec_ton_drs['Authority'] == 'Neath Port Talbot CBC', 
                                                    np.NaN, hhkerb_rec_ton_drs['DRS Aluminium Cans'])
hhkerb_rec_ton_drs['DRS Aluminium Cans'] = np.where(hhkerb_rec_ton_drs['Authority'] == 'Powys County Council', 
                                                    np.NaN, hhkerb_rec_ton_drs['DRS Aluminium Cans'])
hhkerb_rec_ton_drs['DRS Ferrous Cans'] = np.where(hhkerb_rec_ton_drs['Mixed cans'].notnull(), 
                                                    hhkerb_rec_ton_drs['Mixed cans']*.73*.18, 
                                                  hhkerb_rec_ton_drs['DRS Ferrous Cans'])
hhkerb_rec_ton_drs['DRS Ferrous Cans'] = np.where(hhkerb_rec_ton_drs['Authority'] == 'Neath Port Talbot CBC', 
                                                  np.NaN, hhkerb_rec_ton_drs['DRS Ferrous Cans'])
hhkerb_rec_ton_drs['DRS Ferrous Cans'] = np.where(hhkerb_rec_ton_drs['Authority'] == 'Powys County Council', 
                                                  np.NaN, hhkerb_rec_ton_drs['DRS Ferrous Cans'])
hhkerb_rec_ton_drs[['Authority','Aluminium cans','Steel cans','Mixed cans','DRS Aluminium Cans','DRS Ferrous Cans']].head(22)
```

```python
hhkerb_rec_ton_drs.to_csv(op.join(data_dir, 'hhkerb_rec_ton_drs.csv'), encodings = 'utf-8')
```

#Comingled

```python
hhkerb_rec_ton_drs['DRS Aluminium Cans'] = (hhkerb_rec_ton_drs['DRS Aluminium Cans']
                                            .replace(np.NaN, 
                                                     hhkerb_rec_ton_drs['sum_dry_rec']
                                                     *0.01688))
hhkerb_rec_ton_drs[['DRS Aluminium Cans', 'sum_dry_rec']].head(22)
```

```python
hhkerb_rec_ton_drs['DRS Ferrous Cans'] = (hhkerb_rec_ton_drs['DRS Ferrous Cans']
                                            .replace(np.NaN, 
                                                     hhkerb_rec_ton_drs['sum_dry_rec']
                                                     *0.0101))
hhkerb_rec_ton_drs[['DRS Ferrous Cans', 'sum_dry_rec']].head(22)
```

```python
hhkerb_rec_ton_drs['DRS Glass Bottles'] = (hhkerb_rec_ton_drs['DRS Glass Bottles']
                                            .replace(np.NaN, 
                                                     hhkerb_rec_ton_drs['sum_dry_rec']
                                                     *0.1599))
hhkerb_rec_ton_drs[['DRS Glass Bottles', 'sum_dry_rec']].head(22)
```

```python
hhkerb_rec_ton_drs['DRS Plastic Bottles'] = (hhkerb_rec_ton_drs['DRS Plastic Bottles']
                                            .replace(np.NaN, 
                                                     hhkerb_rec_ton_drs['sum_dry_rec']
                                                     *0.0669))
hhkerb_rec_ton_drs[['DRS Plastic Bottles', 'sum_dry_rec']].head(22)
```

```python
hhkerb_rec_ton_drs['DRS Beverage Cartons'] = hhkerb_rec_ton_drs['Composite food and beverage cartons']
hhkerb_rec_ton_drs['DRS Beverage Cartons'] = (hhkerb_rec_ton_drs['DRS Beverage Cartons'].replace(
                                              np.NaN, hhkerb_rec_ton_drs['sum_dry_rec']*0.0031))
hhkerb_rec_ton_drs[['DRS Beverage Cartons', 'sum_dry_rec']].head(22)
```

```python
hhkerb_rec_ton_drs.to_csv(op.join(data_dir, 'hhkerb_rec_ton_drs_2910.csv'), encodings = 'utf-8')
```

```python
hhkerb_waste_ton_drs = pd.DataFrame(columns=['Authority','DRS Glass Bottles',
                                             'DRS Plastic Bottles','DRS Ferrous Cans',
                                             'DRS Aluminium Cans','DRS Beverage Cartons'])
hhkerb_waste_ton_drs['Authority'] = hhkerb_waste_ton_agg['Authority']
hhkerb_waste_ton_drs['DRS Glass Bottles'] = hhkerb_waste_ton_agg['Collected household waste : Regular Collection']*0.0204
hhkerb_waste_ton_drs['DRS Plastic Bottles'] = hhkerb_waste_ton_agg['Collected household waste : Regular Collection']*0.0151
hhkerb_waste_ton_drs['DRS Ferrous Cans'] = hhkerb_waste_ton_agg['Collected household waste : Regular Collection']*0.001554
hhkerb_waste_ton_drs['DRS Aluminium Cans'] = hhkerb_waste_ton_agg['Collected household waste : Regular Collection']*0.003255
hhkerb_waste_ton_drs['DRS Beverage Cartons'] = hhkerb_waste_ton_agg['Collected household waste : Regular Collection']*0.0037
hhkerb_waste_ton_drs
```

```python
hhkerb_waste_ton_sum = hhkerb_waste_ton_drs.sum(axis=0, numeric_only=True).div(1000).to_frame().rename(columns={0: 'Household Kerbside Residual'})
hhkerb_waste_ton_sum
```
