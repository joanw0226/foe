#How to add reuse materials to the recycling stream

```python
hhkerb_rec_ton_qtr = massflow_baseline.get_hhkerb_rec_ton_qtr()
hhkerb_rec_ton_qtr.head()
```

```python
hhkerb_reuse_ton = question_10.groupby('ColText').get_group('Tonnage Collected for Reuse')
hhkerb_reuse_ton_qtr = hhkerb_reuse_ton.pivot_table(values='Data', index=['Authority','Period'],
                                         columns='RowText', aggfunc = lambda x: x)
hhkerb_reuse_ton_qtr = hhkerb_reuse_ton_qtr.reset_index()
hhkerb_reuse_ton_qtr
```

```python
hhkerb_reuse_ton_all = hhkerb_reuse_ton_qtr.groupby('Authority').agg(np.sum)
hhkerb_reuse_ton_all = hhkerb_reuse_ton_all.reset_index()
hhkerb_reuse_ton_all['sum_dry_rec'] = hhkerb_reuse_ton_all.sum(axis=1)
hhkerb_reuse_ton_all
```

```python
hhkerb_rec_ton_all = massflow_baseline.get_hhkerb_rec_ton_la()
hhkerb_rec_ton_all
```

```python
test_merge = hhkerb_rec_ton_all.merge(hhkerb_reuse_ton_all[['Authority','sum_dry_rec']], how='left', on='Authority')
test_merge
```

```python
test_merge['sum_dry_rec_y'] = test_merge['sum_dry_rec_y'].replace(np.NaN, 0)
test_merge['sum_dry_rec'] = test_merge['sum_dry_rec_x'] + test_merge['sum_dry_rec_y']
```

```python
test_merge = test_merge.drop(['sum_dry_rec_x', 'sum_dry_rec_y'], axis=1)
test_merge
```
