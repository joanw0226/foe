```python
res = (raw[(raw.QuestionNumber == 'Q023') & (raw.ColText == 'Tonnage')]
             .pivot_table(values='Data', index=['Authority','Period'],
                          columns='RowText', aggfunc = lambda x: x)
             .reset_index())
res.columns
```

```python
pub_res_qtr = res[['Authority','Period','Collected household waste : Street Cleaning']]
#Only Merthyr did not have any data
```

```python
pub_res_qtr
```

```python
pub_fly_qtr = res[['Authority','Period','Waste Arising from clearance of fly-tipped materials']]
pub_fly_qtr
#not all LAs have this category, but Merthyr does...
```

```python
qu18 = raw[raw['QuestionNumber']=='Q018']
qu18.RowText.unique()
#not relevant i believe
```

```python
pub_res_la = pub_res_qtr.groupby('Authority').agg(np.sum).reset_index()
pub_fly_la = pub_fly_qtr.groupby('Authority').agg(np.sum).reset_index()              
```

```python
merge = pub_res_la.merge(pub_fly_la, how='left',on='Authority')
```

```python
merge
```

```python
merge = merge.replace(np.NaN, 0)
merge['Litter'] = (merge['Collected household waste : Street Cleaning']/2) - merge['Waste Arising from clearance of fly-tipped materials']
```

```python
merge
```

```python
sum(merge['Litter'])
```
