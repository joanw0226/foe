#Applying the composition rates to recycling

```python
hhkerb_rec_ton_agg['DRS Ferrous Cans'] = hhkerb_rec_ton_agg['Steel cans']*.18
hhkerb_rec_ton_agg['DRS Aluminium Cans'] = hhkerb_rec_ton_agg['Aluminium cans']*.80
hhkerb_rec_ton_agg.head()
```

```python
#Using WRAP-calculated rates
hhkerb_rec_ton_agg['DRS Glass Bottles'] = hhkerb_rec_ton_agg['Mixed glass']*.66
```

```python
#Mostly, 'mixed Plastic Bottles' are treated as PET, HDPE, and Other. So DRS is assumed to be only PET and HDPE, 96%
hhkerb_rec_ton_agg['DRS Plastic Bottles'] = hhkerb_rec_ton_agg['Mixed Plastic Bottles']*.96
hhkerb_rec_ton_agg[['Mixed Plastic Bottles','DRS Plastic Bottles']].head(22)
```

```python
# If info not provided from Mixed Plastic Bottles, use Plastics. For most of these LAs, plastics are dense plastics. 
hhkerb_rec_ton_agg['DRS Plastic Bottles'] = np.where(hhkerb_rec_ton_agg['DRS Plastic Bottles'].isnull(), hhkerb_rec_ton_agg['Plastics']*.68, hhkerb_rec_ton_agg['DRS Plastic Bottles'])
hhkerb_rec_ton_agg[['Mixed Plastic Bottles','Plastics', 'DRS Plastic Bottles']].head(22)
```

```python
#For Swansea, Plastics are dense plastics plus plastic film, so the rate is 57% (from dividing 6.69 from 11.66)
hhkerb_rec_ton_agg[hhkerb_rec_ton_agg['Authority'] == 'City and County of Swansea']
```
